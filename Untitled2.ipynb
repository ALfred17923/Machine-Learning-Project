```python
import pandas as pd
```

```python
dataset=pd.read_csv("/content/Interview_Pi_masked_data_filt1.csv")
```

```python
dataset.info()
```

```python
dataset.drop(columns=['YEAR'],inplace=True)
```

```python
dataset.info()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
dataset_enc=pd.get_dummies(dataset)
```

```python
dataset_enc.head()
```

```python
dataset_enc.isnull().sum()
```

```python
dataset_enc.fillna(0,inplace=True)
```

```python
dataset_enc.isnull().sum()
```

```python
dataset_enc=dataset_enc.astype(int)
```

```python
dataset_enc.head()
```

```python
from sklearn.preprocessing import MinMaxScaler,StandardScaler
```

```python
x=dataset_enc.drop(columns=['Commission'])
y=dataset_enc['Commission']
```

```python
print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=43)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
x=scaler.transform(x)
```

Model Building

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr_pred=lr.predict_proba(x_test)
```

```python
y_hat_train=lr.predict(x_train)
y_hat_test=lr.predict(x_test)
```

```python
y_hat_train
```

```python
y_hat_test
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
def model_eval(actual,predicted):
  class_report=classification_report(actual,predicted)
  conf_matrix=confusion_matrix(actual,predicted)
  accuracy=accuracy_score(actual,predicted)

  print("accuracy is",round(accuracy,2))
  print("classification report is",class_report)
  print("confusion matrix is",conf_matrix)
```

```python
model_eval(y_train,y_hat_train)
```

```python
model_eval(y_test,y_hat_test)
```

random forest

```python
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
```

```python
y_hat_train_rf=rf.predict(x_train)
y_hat_test_rf=rf.predict(x_test)
```
