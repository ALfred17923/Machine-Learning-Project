```python
from sklearn.metrics import accuracy_score
```

```python
def model_eval(actual,predicted):
  # Calculate confusion matrix
  confu_matrix = confusion_matrix(actual, predicted)

    # Calculate classification report
  class_report = classification_report(actual, predicted)


  acc_score = accuracy_score(actual, predicted) # Calculate accuracy score
  print("Accuracy: ", round(acc_score, 2))
  print("Confusion Matrix:")
  print(confu_matrix)
  print("Classification Report:")
  print(class_report)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
model_eval(y_train, y_hat_train)
```

```python
model_eval(y_test,y_hat_test)
```
